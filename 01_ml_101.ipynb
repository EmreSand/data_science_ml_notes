### What is Machine Learning:
- a branch of artificial intelligence (AI) and computer science which focuses on the use of data and algorithms to imitate the way that humans learn, gradually improving its accuracy.

#### AI/Machine Learning/Data Science


```AI```:  multidisciplinary field, aiming to artificially create “intelligence” to machines, similar to that displayed by humans. Machines that mimic cognitive functions such as learning and problem-solving.

```Machine Learning```: Subset of AI, scientific study of algorithmic, mathematical, and statistical models which performs a specific task by analyzing data. appproach to try and achieve AI through systems that can find patterns in a set of data

```Deep Learning```: (Deep Neural Networks) Subset of ML One of the techniques to implement ML, consider as a type of Algorithm. teaches computers to process data in a way that is inspired by the human brain. Deep learning models can recognize complex patterns in pictures, text, sounds, and other data to produce accurate insights and predictions

```Data Science```: Extraction of meaningful knowledge and insights from data with the help of various mathematical and Machine Learning-based algorithms, and doing something with it (e.g. Business goal) (Huge overlapping Area with ML, strongly connected)

### Types of ML
- ***Supervised Learning***: Data has already categories. Like csv file with labeled columns, classification can be easily done (cats/dogs). Regression (predicting stock prices, hiring staff based on inputs; age, skills etc. )
- ***Unsupervised Learning***: Data does not have labels/categories.Clustering; machine creating groups. Association Rule Learning, associate different thins to predict. customer purchase behavior.
- ***Reinforcement Learning***: Teaching machine through trial and error. Machine playing a game millions of times, till it perfects it (Skill Acquisition, Real Time Learning)

#### Data Analysis/Machine learning
- D.A. : Looking at a set of data, gaining understanding  of it, finding relations, visualize it.
- M.L. : Analyzes set of data, tries to create a model in order machine to create same pattern.
- Both lies under set of Data Science.
