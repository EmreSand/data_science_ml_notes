```python
# Imports
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_digits # dataset from https://scikit-learn.org/ 
```

```python
# retrieving our data
digits = load_digits()
digits # Data called by function as a dictionary
```

```python
# Converting our data into a DataFrame with features
digits_df = pd.DataFrame(digits["data"], columns=digits["feature_names"]) 
digits_df.head()
```

```python
# Constructing whole dataset by adding target column
digits_df["target"] = digits["target"]
digits_df.head()
```

```python
# Now splitting data to X (features) and y(target)
X = digits_df.drop("target", axis=1)
y = digits_df["target"]
```

```python
# Before using it in the model, splitting the data to training and test data with a random seed 
np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
# Selecting Classification model as RandomForestClassifier
clf = RandomForestClassifier()

# Fitting (teaching) our model with training data
clf.fit(X_train, y_train)

# Evaluating/scoring our models efficiency with test data
clf.score(X_test,y_test)
```
