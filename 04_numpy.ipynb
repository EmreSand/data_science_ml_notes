#### Broadcasting
- describes how NumPy treats arrays with different shapes during arithmetic operations. 
- Subject to certain constraints, the smaller array is “broadcast” across the larger array so that they have compatible shapes. - Broadcasting provides a means of vectorizing array operations so that looping occurs in C instead of Python.

#### Dot Product vs. Element Wise


- row elements of first matrix is multiplied by each column elements of second one by one, then sum of this products creates 1 element in the dot product matrix ...

```python
np.random.seed(0)
mat1 = np.random.randint(10, size=(5,3))
mat2 = np.random.randint(10, size=(5,3))

mat1, mat2
```

```python
mat1 * mat2 # Element-wise multiplication
```

```python
# Dot Product E.G:axb dot bxc = axc, 3x2 dot 2x3, inner dims should match, end product 3x3
np.dot(mat1,mat2)
```

```python
mat3 =np.dot(mat1,mat2.T)

mat1.shape, mat2.shape, mat2.T.shape, mat3.shape
```

```python
mat3
```

```python
arr1 = np.array([[1,2],[4,5], [7,8]])
arr2 = np.array([[8,9,10], [2,2,2]])
arr1, arr2 # 3x2 dot 2x3 = 3x3
# [[1*8+2*2, 1*9+2*2, 1*10+2*2],[4*8+5*2, 4*9+5*2, 4*10+5*2 ],[7*8+8*2, 7*9+8*2, 7*10+8*2]]
```

```python
result= np.dot(arr1,arr2)
result
```

#### Exercise Nut Butter Store Sales

```python
np.random.seed(0)
# number of jars sold through the week
sales_amounts = np.random.randint(20, size=(5,3))
sales_amounts
```

```python
# Create weekly sales dataframe
weekly_sales = pd.DataFrame(sales_amounts,
              index=['Mon', 'Tues', 'Wed', 'Thu', 'Fri'],
              columns=['Almond butter', 'Peanut butter', 'Cashew butter'])
weekly_sales
```

```python
# Create prices array
prices = np.array([10, 8, 12])
prices
```

```python
# Create butter_prices DataFrame
butter_prices = pd.DataFrame(prices, 
                             index=['Price'],
                             columns=['Almond butter', 'Peanut butter', 'Cashew butter'])
butter_prices
```

```python
butter_prices = pd.DataFrame(prices.reshape(1,3), 
                             index=['Price'],
                             columns=['Almond butter', 'Peanut butter', 'Cashew butter'])
butter_prices
```

```python
total_sales = prices.dot(sales_amounts) # tried 3x1 dot 5x3
```

```python
total_sales = prices.dot(sales_amounts.T) 
total_sales
```

```python
# Create daily sales
daily_sales = butter_prices.dot(weekly_sales.T) 
daily_sales
```

```python
# now add a daily total amount column to weekly sale df
weekly_sales["Total ($)"] = daily_sales
weekly_sales # Shape Error, trying to insert 1x5 df to 5x3 df
```

```python
weekly_sales["Total ($)"] = daily_sales.T
weekly_sales
```
