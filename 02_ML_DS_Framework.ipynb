### ML and DS Framework
```
1. Problem definition --> 2. Data --> 3. Evaluation --> 4. Features --> 5. Modeling --> 6. Experimentation
```

#### Types of ML Problems
- Supervised: Binary choice or predicting a number: house prices
- Unsupervised: Predict customer behavior: have inputs but not certain outputs
- Transfer: using a model created before to apply to other model to accomplish

#### Types of Data
- Two types. Structured(Excel file, csv) and Unstructured (Images, Natural Language Text, Transcribed Phone Calls, Videos, Audio files)

- Static data: Data in an Excel or csv file
- Streaming data: constantly changing every time.Stock prices data, changing everyday 

#### Types of Evaluation
- Question of if a ML approach is necessary, practical, accurate

#### Features of Data
- What do we already know about the data: Numerical(age, weight)/Categorical (sex, yes/no info) features 
- In a model as a feature gets more common among samples, it works better

#### Modelling - Splitting Data
- 3 Parts to Modelling: 
    - Choosing and Training a Model
    - Tuning a Model
    - Model Comparison



- Example: Dataset of 100 patients. 70 of them should be used in training, 15 in validation of model, last 15 to test it

#### Modelling - Picking the Model
- Based on our problem and data 
    - Check which model might suit better for problem
    - Check your data type(structured/unstructured)
    - Start small and build up (add complexity) as you need

#### Modelling - Tuning
- ML models have adjustable hyperparameters
- Model's first results are not final
- Tuning can be applied on training or validation data sets


#### Modelling - Comparison


- It has too causes:
    - Data leakage: Test data leaks into training data
    - Data miss-match: Test data is different from the training data

- Possible fixes
    - Using a more advanced model
    - Increasing model hyperparameters
    - Reducing amount of features
    - Training longer
    - Collect more data
    - Try a less advanced model
